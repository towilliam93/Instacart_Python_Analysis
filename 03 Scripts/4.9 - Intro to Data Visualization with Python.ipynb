### Importing Data and Libraries

```python
# Importing Libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Importing data
path = r'C:\Users\arafo\CareerFoundry\Instacart Basket Analysis'
ords_prods_customers = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_customers.pkl'))
```

### Plotting a Bar Chart

```python
# Plotting a Bar Chart
ords_prods_customers['order_days_of_week'].value_counts().plot.bar()
```

```python
# Sorting by index order, using sort_index()
ords_prods_customers['order_days_of_week'].value_counts().sort_index().plot.bar()
```

```python
# sort_index() can also be used when printing frequencies
ords_prods_customers['order_days_of_week'].value_counts()
```

```python
ords_prods_customers['order_days_of_week'].value_counts().sort_index()
```

```python
# Adding colors to the bar chart
bar = ords_prods_customers['order_days_of_week'].value_counts().sort_index().plot.bar(color = ['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Exporting the bar chart
# Call the above bar chart using *name*.figure.
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

# Creating Line Charts

```python
# Seed creation and Sampling
np.random.seed(4)
```

```python
# Create a list holding true/false value to test
dev = np.random.rand(len(ords_prods_customers)) <= 0.7
```

```python
dev
```

```python
# Storing 70% of dataframe in big, and 30% in small. 
big = ords_prods_customers[dev]
```

```python
small = ords_prods_customers[~dev]
```

```python
# Testing for size
len(ords_prods_customers)
```

```python
len(big) + len(small)
```

```python
# Reducing sample size to two columns
df_2 = small[['order_days_of_week','prices']]
```

```python
# Creating a line chart
line = sns.lineplot(data = df_2, x = 'order_days_of_week', y = 'prices')
```
