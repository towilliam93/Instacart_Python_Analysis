### Importing Data and Libraries

```python
# Importing libraries
import pandas as pd
import numpy as np
import os
```

```python
# Importing Data
path = r'C:\Users\arafo\CareerFoundry\Instacart Basket Analysis'
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_derived.pkl'))
```

### Creating a Subset

```python
df = ords_prods_merge[:1000000]
```

```python
df.shape
```

```python
df.head(10)
```

### Grouping with pandas

```python
df.groupby('product_name')
```

### Aggregations

```python
# Aggregation using agg() - Outputting average number of orders per department_id
# Split data into groups based on 'department_id'
# Apply agg() function to each group to obtain mean values for "order_number" column
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Aggregating without agg() - Outputs different format but same info
df.groupby('department_id')['order_number'].mean()
```

### Multiple Aggregations

```python
# Multiple Aggregations using agg()
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

### Aggregating using transform()

```python
# Split the data into groups based on 'user_id' column
# Apply transform() function on 'order_number' to generate max orders per user
# Create a new column, 'max_order' for aggregation results
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.head()
```

### Deriving Columns with loc()

```python
# Deriving Columns with loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Frequency count
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Limiting outputs to select columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

### Tasks

```python
# Aggregation for the entire dataframe
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Ran again in order to more easily compare the two
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# The first difference is that this grabs every 'department_id' column in the dataframe.
# We can also see that department_id 17, household goods, is actually fairly comparable to all the other departments.
```

```python
# Difference between spending habits
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
# Loyal customers typically spend less on products, average around $10.
```

```python
# Targeting different kinds of spenders
# Creating an average price column
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(20)
```

### Creating a Spender Flag

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

### Creating a Frequency Flag

```python
# Frequent vs Non-frequent Customers based on days_since_prior_order
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent Customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
