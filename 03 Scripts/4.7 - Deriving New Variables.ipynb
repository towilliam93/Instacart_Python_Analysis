### Creating a Price Range Flag

```python
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

### Alternative Ways of Creating Price Range Flag

```python
# If Statements using loc() function
# Use different cells for each condition
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

### Creating Busiest Days Flag

```python
df_ords_prods_merge['order_days_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in df_ords_prods_merge["order_days_of_week"]:
    if value == 0:
        result.append("Busiest Day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### Tasks

```python
# Busiest Days column
# Values taken from previous count
result2 = []

for value in df_ords_prods_merge["order_days_of_week"]:
    if value == 0 or value == 1:
        result2.append("Busiest Day")
    elif value == 4 or value == 3:
        result2.append("Slowest Day")
    else:
        result2.append("Regularly busy")
```

```python
df_ords_prods_merge['busiest_days'] = result2
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

```python
# Busiest days were 0 and 1, or Saturdays and Sundays, with 6204182 and 5660230 orders, totalling 11864412
# Least busy days were 4 and 3, or Wednesdays and Tuesdays, with 3783802 and 3840534 orders, totalling 7624336
```

```python
# Busiest Hours
```

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Based off this frequency, we can see about a 400,000 gap between hours 9 and 17, and another gap between hours 6 and 23.
result3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value >= 9 and value <= 16:
        result3.append("Most Orders")
    elif value >= 0 and value <= 6:
        result3.append("Fewest Orders")
    else:
        result3.append("Average Orders")
```

```python
df_ords_prods_merge['busiest_period_of_day'] = result3
```

```python
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
